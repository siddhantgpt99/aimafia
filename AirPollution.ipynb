```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

training = pd.read_csv('Train.csv')



feature_1 = training['feature_1'].values
feature_2 = training['feature_2'].values
feature_3 = training['feature_3'].values
feature_4 = training['feature_4'].values
feature_5 = training['feature_5'].values
target = training['target'].values

feature_1 = feature_1.reshape((-1,1))
feature_2 = feature_2.reshape((-1,1))
feature_3 = feature_3.reshape((-1,1))
feature_4 = feature_4.reshape((-1,1))
feature_5 = feature_5.reshape((-1,1))
target = target.reshape((-1,1))

Feature_1 = (feature_1-feature_1.mean())/feature_1.std()
Feature_2 = (feature_2-feature_2.mean())/feature_2.std()
Feature_3 = (feature_3-feature_3.mean())/feature_3.std()
Feature_4 = (feature_4-feature_4.mean())/feature_4.std()
Feature_5 = (feature_5-feature_5.mean())/feature_5.std()
Target = target
```

```python
plt.scatter(Feature_2,Target)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
model2 = LinearRegression()
model3 = LinearRegression()
model4 = LinearRegression()
model5 = LinearRegression()
```

## Training

```python
model1.fit(Feature_1,Target)
model2.fit(Feature_2,Target)
model3.fit(Feature_3,Target)
model4.fit(Feature_4,Target)
model5.fit(Feature_5,Target)
```

# Predictions

```python
test = pd.read_csv('Test.csv')

feature_1n = test['feature_1'].values
feature_2n = test['feature_2'].values
feature_3n = test['feature_3'].values
feature_4n = test['feature_4'].values
feature_5n = test['feature_5'].values

feature_1n = feature_1n.reshape((-1,1))
feature_2n = feature_2n.reshape((-1,1))
feature_3n = feature_3n.reshape((-1,1))
feature_4n = feature_4n.reshape((-1,1))
feature_5n = feature_5n.reshape((-1,1))



Feature_1n = (feature_1n-feature_1n.mean())/feature_1n.std()
Feature_2n = (feature_2n-feature_2n.mean())/feature_2n.std()
Feature_3n = (feature_3n-feature_3n.mean())/feature_3n.std()
Feature_4n = (feature_4n-feature_4n.mean())/feature_4n.std()
Feature_5n = (feature_5n-feature_5n.mean())/feature_5n.std()



output1 = model1.predict(Feature_1n)
output2 = model2.predict(Feature_2n)
output3 = model3.predict(Feature_3n)
output4 = model4.predict(Feature_4n)
output5 = model5.predict(Feature_5n)
```

```python
# Parameters Learnedm
```

```python
bias1 = model1.intercept_
coeff1 = model1.coef_

bias2 = model2.intercept_
coeff2 = model2.coef_

bias3 = model3.intercept_
coeff3 = model3.coef_

bias4 = model4.intercept_
coeff4 = model4.coef_

bias5 = model5.intercept_
coeff5 = model5.coef_


print(bias1,bias2,bias3,bias4,bias5)
print(coeff1,coeff2,coeff3,coeff4,coeff5)
```

## Score 

```python
model1.score(Feature_1,Target), model2.score(Feature_2,Target), model3.score(Feature_3,Target), model4.score(Feature_4,Target), model5.score(Feature_5,Target)
```
